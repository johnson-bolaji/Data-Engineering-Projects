### **Bing News Data Analytics- End to End Azure Data Engineering Project using Microsoft Fabric**
This project involves ingesting the latest news data from Bing API, transforming the data into a structured Delta table, performing sentiment analysis using a text analytics model, creating a news dashboard using Power BI, and setting up alerts using Data Activator. The project demonstrates the collaboration and orchestration capabilities of Microsoft Fabric, as well as the use of incremental load and data activator concepts.
